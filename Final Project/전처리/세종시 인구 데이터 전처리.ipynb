```python
import pandas as pd
```

```python
# 전체 데이터
df = pd.read_excel('C:/finalproject/기관별연령별인구통계(1세단위 23.9.30. 기준).xlsx')
df
```

```python
# 나이
age = df[5:].iloc[4::3]['Unnamed: 1'].reset_index(drop=True)
age
```

```python
# 총인구수
total_pop = df[5:].iloc[4::3, 3].reset_index(drop=True)
total_pop
```

```python
# 남자 인구수
m_pop = df[5:].iloc[5::3,3].reset_index(drop=True)
m_pop
```

```python
# 여자 인구수
f_pop = df[5:].iloc[6::3,3].reset_index(drop=True)
f_pop
```

```python
# 연령별 남자, 여자 인구수
population = pd.concat([age, total_pop, m_pop, f_pop], axis=1)
population.columns = ['age','total', 'm', 'f']
population
```

```python
population.to_csv('C:/finalproject/project_population.csv')
```
