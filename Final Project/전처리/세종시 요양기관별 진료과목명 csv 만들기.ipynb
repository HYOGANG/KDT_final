```python
import pandas as pd
```

```python
df = pd.read_excel('5.의료기관별상세정보서비스_03_진료과목정보.xlsx')
df = df.iloc[:][['요양기관명', '진료과목코드명']]
df
```

```python
df1 = pd.read_excel('1.병원정보서비스.xlsx')
df1 = df1.iloc[:][['시도코드명', '요양기관명','종별코드명']]
df1
```

```python
# 세종시 요양기관 추출
df1 = df1.query('시도코드명 == "세종시"')
df1
```

```python
# 세종시 요양기관별 진료과목코드 추출
df2 = pd.merge(df1, df, how='left', on='요양기관명')
df2 = df2.drop_duplicates()
df2
```

```python
df2.isnull().sum()
```

```python
df2[df2.진료과목코드명.isna( )]
```

```python
# 요양기관의 진료과목코드 최종
df3 = df2
df3['진료과목코드명'] = df3.groupby(['시도코드명','요양기관명','종별코드명']).tran
sform('sum')
df3 = df3.drop_duplicates()
df3['진료과목코드명'] = df3['진료과목코드명'].str.replace('과', '과,')
df3
```

```python
df3.to_csv("project_hos_department.csv")
```
