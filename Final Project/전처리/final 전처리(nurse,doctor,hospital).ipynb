```python
import pandas as pd
```

```python
h=pd.read_excel("C:/Users/user/Final Project/원데이터/hospital_count.xlsx", header =None)
n=pd.read_excel("C:/Users/user/Final Project/원데이터/nurse.xlsx", header =None)
d=pd.read_excel("C:/Users/user/Final Project/원데이터/doctor.xlsx", header =None)
```

```python
df=h.iloc[2:4,3:].T
df.columns=['type','hospital']
df
```

```python
# count object에서 int 형태로 변환
df['hospital'] = df['hospital'].astype('int')
print(df.dtypes)
```

```python
# 간호사

df1=n.iloc[5:,1:3]
df1=df1.rename(columns=df.iloc[0])  
df1.columns=['type','nurse']              # 컬럼명 변경 
df1
```

```python
# 하이픈 0로 변경
df1=df1.replace(to_replace={'-':'0'})
df1
```

```python
# count object에서 int 형태로 변환
df1['nurse'] = df1['nurse'].astype('int')
print(df1.dtypes)
```

```python
# 의사

df2=d.iloc[5:,[1,2,7,12,18]]
df2.columns=['type','의사','치과의사','한의사','약사']
df2=df2.replace(to_replace={'-':'0'})
df2
```

```python
df2.dtypes
df2 = df2.astype({'의사':'int','치과의사':'int','한의사':'int' ,'약사':'int'})
```

```python
df2["doctor"] = df2.iloc[:,1:].sum(axis=1)
df2
```

```python
df3=df2.iloc[:,[0,5]]
df3
```

```python
test = pd.merge(left=df, right=df1, how ="outer", on="type")
test
```

```python
final=pd.merge(left=test, right = df3, how = "outer", on="type")
final
```

```python
final = final.fillna(0)
final
```
